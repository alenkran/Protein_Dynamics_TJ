```python
import numpy as np
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Load PCA coordinate from ISOMAP
file_location = 'C:\\Users\\Christian\\Documents\\GitHub\\Protein_Dynamics_TJ\\MSM_Builder_Test\\Sherlock_Data\\'
file_name = ['40_10_1.0_4', '30_40_1.0_4', '40_2_1.0_2'] #fspeptide
for i, ID in enumerate(file_name):
    plt.figure(i)
    pca = np.load(file_location + 'X_pc_isomap_'+ ID + '.dat')
    plt.hexbin(pca[:,0], pca[:,1], bins='log', mincnt=1, cmap='viridis')
    plt.title(ID)
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
```

```python
# Load PCA coordinate from ISOMAP
file_location = 'C:\\Users\\Christian\\Documents\\GitHub\\Protein_Dynamics_TJ\\MSM_Builder_Test\\PC_Data\\'
method = 'X_pc_isomap'
#method = 'X_pc_kernelPCA'
parameter = '_30_10_'
file_name = np.arange(250, 1050, 50).astype(str)
for i, ID in enumerate(file_name):
    plt.figure(i)
    pca = np.load(file_location + method + parameter + ID + '_4.dat')
    try:
        plt.hexbin(pca[:,0], pca[:,1], gridsize=100, bins='log', mincnt=1, cmap='gray_r')
    except:
        print "ID" + " failed"
    plt.title(method+ID)
    plt.xlabel('PCA 1')
    plt.ylabel('PCA 2')
    cb = plt.colorbar()
    cb.set_label('log10(N)')
    plt.savefig(method+parameter + ID + '_4.png', bbox_inches='tight')
```
