```python
# Load PCA coordinate from ISOMAP
file_location = 'C:\\Users\\Christian\\Documents\\GitHub\\Protein_Dynamics_TJ\\MSM_Builder_Test\\PC_Data\\'
method_list = ['X_pc_isomap','X_pc_kernelPCA']
parameter = '_30_10_'
file_name = np.arange(5000, 16000, 1000).astype(str)

for method in method_list:
    for i, ID in enumerate(file_name):
        plt.figure()
        try:
            pca = np.load(file_location + method + parameter + ID + '_4.dat')
        except:
            print "ID" + " failed"
            
        x = pca[:,0]
        y = pca[:,1]
        xedges, yedges = np.linspace(min(x), max(x), 80), np.linspace(min(y), max(y), 80)
        hist, xedges, yedges = np.histogram2d(pca[:,0],pca[:,1],(xedges, yedges))
        hist = np.log(hist)
        xidx = np.clip(np.digitize(pca[:,0], xedges), 0, hist.shape[0]-1)
        yidx = np.clip(np.digitize(pca[:,1], yedges), 0, hist.shape[1]-1)
        c = hist[xidx, yidx]
        plt.scatter(pca[:,0], pca[:,1], c=c, edgecolor='none')
        cb = plt.colorbar()
        cb.set_label('log10(N)')
        
        extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
        #cset1 = plt.contour(np.transpose(hist),extent=extent)
        
        plt.title(method+ID)
        plt.xlabel('PCA 1')
        plt.ylabel('PCA 2')
        #plt.savefig(method+parameter + ID + '_4.png', bbox_inches='tight')
```
