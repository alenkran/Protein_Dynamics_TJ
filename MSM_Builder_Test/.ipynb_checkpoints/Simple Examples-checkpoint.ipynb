# Toy Example 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import manifold, datasets
from sklearn.decomposition import PCA
%matplotlib inline
```

### Most obvious example, structures lie on a manifold

```python
n_points = 1000
X, color = datasets.samples_generator.make_s_curve(n_points, random_state=0)
X = X[:, [0,2]] + np.random.normal(scale=0.2, size=(1000, 2))
n_components = 2
```

```python
plt.scatter(X[:, 0], X[:, 1] , c=color, cmap=plt.cm.Spectral)
```

```python
# Regular PCA
pca = PCA(n_components=2)
pca_X = pca.fit_transform(X)

# ISOMAP/PCA
n_neighbors = 20
isomap = manifold.Isomap(n_neighbors=n_neighbors, n_components=n_components)
isomap_pca_X = pca.fit_transform(isomap.fit_transform(X))
```

```python
plt.figure(figsize(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(pca_X[:, 0], np.random.normal(scale=0.001, size=(1000, 1)) , c=color, cmap=plt.cm.Spectral)
xlabel('PC 1, important direction found by PCA'); plt.title('Dynamical Direction found by PCA')
plt.subplot(1, 2, 2)
plt.scatter(isomap_pca_X[:, 0], np.random.normal(scale=0.001, size=(1000, 1)) , c=color, cmap=plt.cm.Spectral)
xlabel('ISOMAP/PC 1'); plt.title('Dynamical Direction found by ISOMAP/PCA')
```

```python
# ISOMAP/PCA
n_neighbors = 10
isomap = manifold.Isomap(n_neighbors=n_neighbors, n_components=n_components)
isomap_pca_X = pca.fit_transform(isomap.fit_transform(X))
```

```python
plt.scatter(isomap_pca_X[:, 0], np.random.normal(scale=0.001, size=(1000, 1)) , c=color, cmap=plt.cm.Spectral)
xlabel('ISOMAP/PC 1'); plt.title('Dynamical Direction found by PCA')
```
